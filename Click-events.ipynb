```python
from ipywidgets import widgets
import numpy as np
import matplotlib.pyplot as plt

# returns figure object
click_area = plt.subplots()[0]
click_area.canvas.set_window_title('Click here')
text_area = widgets.HTML()

def onclick(event):
    text_area.value = 'xdata=%f, ydata=%f'%(event.xdata, event.ydata)

click_area.canvas.mpl_connect('button_press_event', onclick)
display(text_area)
```

```python
from ipywidgets import widgets
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import IFrame
import os

filepath = os.path.realpath('./modified.html')
with open(filepath, 'r') as f:
    htmlsample = f.read()
    
# An HTML file is read and is stored in the variable `htmlsample`
# Equivalent to
# htmlsample = "<html><body>Click here</body></html>"

click_area = widgets.HTML(htmlsample)
text_area = widgets.HTML("Co-ordinates will be displayed here")

def onclick(event):
    text_area.value = 'xdata=%f, ydata=%f'% (event.xdata, event.ydata)

click_area.observe(onclick)

display(text_area, click_area)
```
